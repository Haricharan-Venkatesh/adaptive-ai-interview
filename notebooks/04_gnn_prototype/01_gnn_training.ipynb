# GNN Competency Propagation Prototype

This notebook demonstrates how the Graph Attention Network (GAT) propagates mastery estimates across a prerequisite graph.

```python
import sys
import os
sys.path.append(os.path.abspath('../../backend'))

from ai_models.knowledge_tracing.gnn import GNNKnowledgePropagator
```

```python
# Initialize propagator for a small graph with 5 skills
propagator = GNNKnowledgePropagator(num_skills=5)

# Edges (prerequisite -> dependent)
# 0 -> 1 -> 2 -> 3 -> 4
edges = [[0, 1, 2, 3], [1, 2, 3, 4]]

# Initial mastery estimates (skill 0 is mastered, others are priors)
mastery = [0.95, 0.5, 0.5, 0.5, 0.5]

outputs = propagator.propagate(mastery, edges)
print("Mastery probabilities:\n", outputs['mastery_probs'].squeeze().tolist())
print("Complexity scores:\n", outputs['complexity_scores'].squeeze().tolist())
```
